```python
def generate_input_text(row):
    # Mappings
    gender = gender_map.get(row["Gender"], "unknown")
    marital = marital_map.get(row["MaritalStatus"], "unknown")
    work_type = work_type_map.get(row["PartTimeFullTime"], "unknown")

    # Values for formatting
    values = {
        "Age": row["Age"],
        "Gender": gender,
        "MaritalStatus": marital,
        "DateOfAccident": row["DateOfAccident"],
        "ReportedDay": row["ReportedDay"],
        "PartTimeFullTime": work_type,
        "HoursWorkedPerWeek": int(row["HoursWorkedPerWeek"]),
        "DependentChildren": int(row["DependentChildren"]),
        "DependentsOther": int(row["DependentsOther"]),
        "WeeklyRate": int(row["WeeklyRate"]) if row["WeeklyRate"] > 0 else "not specified"
    }

    # Candidate sentences list
    parts = []

    # Randomly choose 2 to 5 columns (excluding the claim)
    available_keys = list(templates.keys())
    selected_keys = random.sample(available_keys, k=random.randint(2, 5))

    # Generate sentence for each selected feature
    for key in selected_keys:
        value = values[key]
        template = random.choice(templates[key])
        sentence = template.format(**{key: value})
        parts.append(sentence)

    # Add the detailed claim as just another sentence
    parts.append(row["DetailedClaimDescription"].strip())

    # Shuffle the whole sentence list to make order random
    random.shuffle(parts)

    # Join and return
    return " ".join(parts)
```

applying function to whole dataset

```python
# Apply the function to the full DataFrame
df["InputText"] = df.apply(generate_input_text, axis=1)

# Save the updated DataFrame to CSV
df.to_csv("claims_3.csv", index=False)

print("✅ InputText column created and saved to claims_3.csv")
```

```python
import random

def add_severity_phrase(text, severity_label):
    phrase = random.choice(severity_templates[severity_label])
    
    # Insert at end with a separating space
    return text.strip() + " " + phrase
```

```python
df["InputText"] = df.apply(
    lambda row: add_severity_phrase(row["InputText"], row["SeverityLabel"]),
    axis=1
)
```
